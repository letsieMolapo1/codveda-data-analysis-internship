```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

Dataframe = pd.read_csv("./cleaned_train.csv")
Dataframe.head()
```

```python
Dataframe.describe()
```

```python
x = Dataframe["Survived"]
plt.hist(x)
plt.xticks(rotation=90 , size = 10)
plt.title("Survival rate")
plt.xlabel("Survived")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize = (10,6))
sns.histplot(data=Dataframe, x = "Fare" , color = "red")
plt.title("Fare Distribution")
plt.xticks(rotation= 90 , size = 12)
plt.xlabel("Fare")
plt.savefig("Fare distribution - histogram")
plt.show()
```

```python
Dataframe["Embarked"].value_counts().sum()
```

```python
sns.scatterplot(data=Dataframe, x = 'Age', y = 'Fare')
plt.title("Fare vs Age")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.savefig("Fare Vs Age - scatterplot")
plt.show()
```

```python
sns.boxplot( data = Dataframe , x = "Sex" , y = "Age")
plt.title("Age distribution by Gender")
plt.xlabel("Gender")
plt.ylabel("Age")
plt.savefig("Age distribution by Gender - boxplot")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(Dataframe.corr(numeric_only = True) , annot = True , cmap="coolwarm" )
plt.title("Correlation Heatmap")
plt.savefig("Correlation Heatmap between numeric values")
plt.show()
```
