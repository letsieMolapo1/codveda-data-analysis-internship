```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

Dataframe = pd.read_csv("./cleaned_train.csv")
Dataframe.head()
```

```python
Fare_by_Gender = Dataframe.groupby("Sex")["Fare"].sum()
plt.figure(figsize=(8,6))
plt.barh( y = Fare_by_Gender.index , width= Fare_by_Gender.values )
plt.title("Gender with most Fares")
plt.xlabel("Fare")
plt.ylabel("Gender")
plt.savefig("Gender with most Fares - Horizontal Bar")
plt.show()
```

```python
sns.lineplot(data=Dataframe, x= 'Fare', y= 'Age', color='green')
plt.title("Age group Fare distribution")
plt.xlabel("Numbers of Fares")
plt.ylabel("Age group")
plt.savefig("Age group Fare distribution - Line chart")
plt.show()
```

```python
sns.barplot(data=Dataframe , x= 'Pclass', y= 'Survived' , color='red')
plt.title("Passenger per class survival rate")
plt.xlabel("Passenger per class")
plt.ylabel("Survival rate")
plt.savefig("Passenger per class survival rate - Barplot")
plt.show()
```

```python
sns.scatterplot(data=Dataframe , x='Fare' , y='Embarked' , hue='Fare')
plt.legend( title = 'Fares')
plt.title("Embarked vs Fare")
plt.xlabel("Fare frequency")
plt.ylabel("Embarked")
plt.savefig("Ebarked vs fare - scatterplot")
plt.show()
```
